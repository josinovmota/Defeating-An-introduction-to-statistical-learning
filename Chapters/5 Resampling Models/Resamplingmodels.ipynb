## Cross-Validation:

Sempre que vamos criar um modelo, precisamos analisar os dados afim de entender qual modelo se encaixa melhor. De incio, usamos dados de treino para treinar o nosso modelo e,  partir das predictions feitas nesses dados de treino, fazemos uma prediction com os dados de teste. Isso faz com que nosso modelo seja bom para dos dados de treino, ou seja, tenha um BAIXO TRAINING ERROR RATE, porm, pode ser que os dados que venham a seguir no tenham as mesmas caracterstica dos de treino, tornando assim, muitas vezes, nosso modelo pior para os dados de TESTE, causando um alto TEST ERROR RATING.
Dado isso, a cross-validation divide os dados de treino em training set e **VALIDATION SET** e faz resampling com uma das partes divididas, fazendo assim, com que tenhamos certeza que nosso modelo tem um bom desempenho mesmo com a variabilidade dos dados

### Leave-One-Out Cross-Validation:

 um dos mtodos de cross-validation aonde, ao invs de dividir nossos dados em TRAINING SET e VALIDATION SET, refazemos nosso modelo sempre retirando 1 AMOSTRA e prevendo ela de forma singular. Ele requer mais tempo de execuo na maioria dos casos, porm, tende a lidar muito bem com **BIAS**, j que eventualmente usar como TREINO cada um dos x e y.

### K-Fold Cross-Validation:

Um dos motivos para usarmos o K-Fold CV no lugar do LOOCV  que o K-Fold requer menos tempo, ou seja, menos trabalho computacional. Em alguns tipos de regresses existe uma frmula que diminui drasticamente esse tempo, porm, em grandes modelos, fazer 1 por 1, como  o caso do LOOCV, requer muito tempo. Ento, por mais que o Leave-One-Out possa gerar algo mais preciso, ele sempre caminha ao lado do K-Fold.

O K dentro do K-Fold significa a quantidade de grupos, ou seja, se o K = 10, dividimos os dados em 10 grupos de tamanho igual

Quando plotamos os mtodos de Cross-Validation em relao  flexibilidade, sabemos que vai existir um ponto mnimo, aonde o MSE  o menor possvel e logo aps esse ponto, o MSE vai comear a subir. #### Bias-Variance Trade-Off for k-fold cross-validation:

K-fold CV tem uma vantagem computacional em relao ao LOOCV, pois o mesmo faz menos iteraes dentro do **N**. Enquanto o LOOCV faz N iteraes, o K-Fold CV faz K iteraes aonde K < N

O Bias-Variance Trade-Off nesse caso nos mostra que o LOOCV tende a ter uma **VARIANCE** maior, porm um **BIAS** menor.  sabido que um modelo bom  aquele modelo que tem POUCA **VARIANCE** e POUCO **BIAS**, como  mostrado a seguir:

### Bootstrap

O processo de bootstrapping nos ajuda a calcular algumas estatsticas de datasets, como **Standard Error**, **Standard Deviation**, sem precisar de dados reais. Nesse processo fazemos um **Resampling** do dataset com replacement, ou seja, tiramos uma amostra dos dados, que podem ser repetidas, calculamos o que buscamos, criamos um novo bootstrap data set com os dados reais e repetimos esse processo vrias vezes.  partir de vrios testes, foi descoberto que esses clculos por meio do bootstrapping do valores muito prximos aos que calculamos usando datasets anteriores.
